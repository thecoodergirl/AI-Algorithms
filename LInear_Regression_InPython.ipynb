```python
#load libraries to collab
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import *
import statsmodels.formula.api as smf
from random import randint
#creating dataset of 1 to 100
x = range(1,101)
#generating random gasoline prices
np.random.seed(2)
y=np.random.randint(low=40, high=60, size=(100,))
y1= np.random.normal(40, 60, 100)/x
Dataset=pd.DataFrame({'Observations' : x, 'Gasoline':y})
Dataset
Dataset.corr()
#plot a scatter diagram
plt.xlabel('Obesrvations')
plt.ylabel('Gasoline Price')
plt.title('Data Distribution')
plt.ylim(0,100)
plt.plot(x, y, 'o') 
#find the intercept and weight
m, b = np.polyfit(x, y, 1)


print('slope = ',m,"and intercept =",b)


#Plotting data and regression line
plt.xlabel('Observations')
plt.ylabel('Gasoline Price')
plt.title('Linear Regression Modelling')
plt.ylim(0,100)
plt.plot(x, y, 'o') 
plt.plot(x, m*x + b) 
plt.show()



#Prediction
x2 = range(101,151)
y1 = m*x2 + b

#Plotting prediction
plt.xlabel('Observations')
plt.ylabel('Gasoline Price')
plt.title('Linear Regression Prediction')
plt.ylim(0,100)
plt.plot(x, y, 'o',label="Data") 
plt.plot(x2, y2,label="Predicted") 
plt.legend(loc="upper right")
plt.show()
```
